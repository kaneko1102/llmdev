```python
# 整数の代入
a = 10
b = -5
c = 0

# 整数同士の計算
sum_result = a + b
product_result = a * b

print("a:", a)
print("b:", b)
print("c:", c)
print("合計:", sum_result)
print("積:", product_result)
```

```python
# 除算の例
division_result = a / 3
integer_division_result = a // 3

print("除算結果:", division_result)
print("整数除算結果:", integer_division_result)
```

```python
# 浮動小数点数の代入
x = 3.14
y = -0.5
z = 2.0 # 小数点をつけると float 型になる

# 浮動小数点数の計算
sum_result = x + y
product_result = x * y

print("x:", x)
print("y:", y)
print("z:", z)
print("合計:", sum_result)
print("積:", product_result)
```

```python
# 浮動小数点数の誤差の例
result = 0.1 + 0.2
print("0.1 + 0.2 =", result)
```

```python
# 複素数の代入
c1 = 2 + 3j
c2 = -1 + 0.5j

# 複素数の計算
sum_result = c1 + c2
product_result = c1 * c2

print("c1:", c1)
print("c2:", c2)
print("合計:", sum_result)
print("積:", product_result)
```

```python
# 文字列の結合
text1 = "Hello"
text2 = "World"
result = text1 + " " + text2
print(result)

# 文字列の繰り返し
text = "Hi! "
print(text * 3)

# 文字列の長さを取得
length = len("Hello, World!")
print("文字列の長さ:",length)
```

```python
# 大文字や小文字に変換
text = "Python"
print(text.upper())
print(text.lower())

# 部分文字列を置換
text = "Hello, World"
print(text.replace("World", "Python"))

# 前後の空白や指定した文字を削除
text = "  Hello, World!  "
print(text.strip())

# 部分文字列の最初の位置を検索
text = "Hello, World!"
print(text.find("World"))
```

```python
# 小数点以下の桁数
pi = 3.14159
print(f"小数点以下の桁数: {pi:.2f}")

# カンマ区切り
large_number = 1234567890
print(f"カンマ区切り: {large_number:,}")

# 文字列の出力幅
name = "Alice"
print(f"名前: {name:10}")

# ゼロ埋め
num = 42
print(f"ゼロ埋め: {num:05}")

# パーセント表示
accuracy = 0.856
print(f"パーセント表示: {accuracy:.1%}")

# 揃え
name = "Alice"
print(f"左揃え: {name:<10}!")
print(f"右揃え: {name:>10}!")
print(f"中央揃え: {name:^10}!")

# 日時のフォーマット
from datetime import datetime
now = datetime.now()
print(f"現在日時: {now:%Y-%m-%d %H:%M:%S}")
```
