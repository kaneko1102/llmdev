```python
# 算術演算子の例
a = 10
b = 3

sum_result = a + b
diff_result = a - b
product_result = a * b
quotient_result = a / b
int_quotient_result = a // b
remainder_result = a % b
power_result = a ** b

print("加算: ", sum_result)
print("減算: ", diff_result)
print("乗算: ", product_result)
print("割算（浮動小数点）: ", quotient_result)
print("整数除算: ", int_quotient_result)
print("余り: ",remainder_result)
print("累乗: ", power_result)
```

```python
# 複合代入演算子の例
x = 10
x += 5
print("x += 5: ", x)

x -= 3
print("x -= 3: ", x)

x *= 2
print("x *= 2: ", x)

x /= 4
print("x /= 4: ", x)

x //= 5
print("x //= 5: ", x)

x %= 3
print("x %= 3: ", x)

x **= 3
print("x **= 3: ",x)
```

```python
# 比較演算子の例
a = 10
b = 5

print("a > b: ", a > b)
print("a < b: ", a < b)
print("a >= b: ", a >= b)
print("a <= b: ", a <= b)
print("a == b: ", a == b)
print("a != b: ", a != b)
```

```python
# 連続比較の例
x = 150
print("100 < x < 200: ", 100 < x < 200)

y = 50
print("100 < y < 200: ", 100 < y < 200)
```

```python
# 論理演算子の例
x = 15
print("10 < x and x < 20: ", 10 < x and x < 20)
print("x < 10 or x > 20: ", x < 10 or x > 20)
print("not ( x == 15): ", not (x == 15))
```

```python
# ビット演算子の例
a = 0b1100 # 12
b = 0b1010 # 10

print("a & b: ", bin(a & b))
print("a | b: ", bin(a | b))
print("a ^ b: ", bin(a ^ b))
print("a << 1: ", bin(a << 1))
print("a >> 1: ", bin(a >> 1))
```

```python
# 各データ型に対する演算子の例
a = 10
b = 3.5
c = True
d = "hello"
e = [1, 2, 3]

# 算術演算子
print(a + 5)
print(b * 2)
print(d + " world")
print(e * 2)

# 比較演算子
print(a > 5)
print(d == "hello")
print(c == True)

# 論理演算子
print(c and False)
print(not c)

# メンバーシップ演算子
print(2 in e)
print("o" in d)

f = e
print(f is e)
```
