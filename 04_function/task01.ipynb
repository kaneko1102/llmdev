```python
def is_valid_length(username):
    if 5 <= len(username) <= 15:
        return True
    else:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        return False


def without_whitespace(username):
    white_space = [" ", "\t", "　"] # 半角スペース、タブ、全角スペース
    for i in white_space:
        if i in username:
            print("エラー: ユーザー名に空白を含めることはできません")
            return False
    return True

def is_valid_alphabet_and_number(username):
    if username.isalnum():
        return True
    else:
        print("エラー: ユーザー名は英数字のみ使用できます")
        return False

def is_valid_username(username):
    if is_valid_length(username) and without_whitespace(username) and is_valid_alphabet_and_number(username):
        return True
    else:
        return False

username = input("ユーザ名を入力してください: ")
if is_valid_username(username):
    print("ユーザー名は有効です")
else:
    print("不正なユーザー名です")
```
