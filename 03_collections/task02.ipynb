```python
word_list = []
char_count_dict = {}

while True:
    word = input("英単語を入力してください：")
    if word == "":
        break
    word_list.append(word)

for word in word_list:
    for char in word:
        if char in char_count_dict:
            char_count_dict[char] += 1
        else:
            char_count_dict[char] = 1


word_list.sort()
print("入力した英単語：", word_list)

char_count_list = sorted(char_count_dict.items())

for i in char_count_list:
    print(f"{i[0]}が{i[1]}個ありました")
```
